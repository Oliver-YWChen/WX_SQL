```python
import openpyxl
from openpyxl import Workbook

book = openpyxl.load_workbook('711.xlsx')
sheet = book.active


T = []


for row in sheet.iter_rows(min_row = 1, min_col = 1, max_row = 5248, max_col = 3):
    n = 0
    t = []
    
    for cell in row:
        n = n + 1
        c = cell.value
        t.append(c)
        
        if n == 3:
            if c[2] == '市':
                a = c.split('市', 1)
                county = a[0] + '市'
                if '區' in a[1]:
                    b = a[1].split('區')
                    town = b[0] + '區'
                if '鄉' in a[1]:
                    b = a[1].split('鄉')
                    town = b[0] + '鄉'
                if '鎮' in a[1]:
                    b = a[1].split('鎮')
                    town = b[0] + '鎮'
                if '市' in a[1]:
                    b = a[1].split('市')
                    town = b[0] + '市'
                
            if c[2] == '縣':
                a = c.split('縣', 1)
                county = a[0] + '縣'
                if '區' in a[1]:
                    b = a[1].split('區')
                    town = b[0] + '區'
                if '鄉' in a[1]:
                    b = a[1].split('鄉')
                    town = b[0] + '鄉'
                if '鎮' in a[1]:
                    b = a[1].split('鎮')
                    town = b[0] + '鎮'
                if '市' in a[1]:
                    b = a[1].split('市')
                    town = b[0] + '市'
                
            t.append(county)
            t.append(town)
            T.append(tuple(t))
    
print(T)
```
