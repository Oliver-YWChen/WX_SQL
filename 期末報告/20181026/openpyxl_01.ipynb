# 此程式示範如何利用Openpyxl產生一個EXCEL檔案，包括：
1. 創建一個工作簿(workbook)
2. 將資料填入工作表(worksheet)
3. 從工作表(worksheet)讀取資料
4. 儲存工作簿(workbook)

```python
from openpyxl import Workbook
import random

# create a new workbook
wb = Workbook()

# grab the active worksheet
ws = wb.active

# create the header cells
ws['A1'] = "Q1"
ws['B1'] = "Q2"
ws['C1'] = "Q3"
ws['D1'] = "Q4"
ws['E1'] = "Total"
ws['F1'] = "Average"

# enter data to the worksheet
for i in range(10):
    t = []
    
    # create a list of 4 random integers between 60 and 100
    for j in range(4):
        g = random.randint(60,100)
        t.append(g)
    
    # enter data to the worksheet by appending the list
    ws.append(t)
    
# read data from the worksheet
row_count = ws.max_row
for row in range(2,row_count+1):
    t = []
    
    # read data from each cell of the worksheet using cell name
    for column in 'ABCD':
        cell_name = "{}{}".format(column, row)
        
        t.append(ws[cell_name].value)
        
    # calculate the total and average
    total = sum(t)
    average = float(total) / len(t)
    
    # insert data into the worksheet using cell name
    total_cell = 'E' + str(row)
    ws[total_cell] = total
    average_cell = 'F' + str(row)
    ws[average_cell] = average
    
# save the workbook    
wb.save('test01.xlsx')
```
