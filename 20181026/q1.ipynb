```python
from openpyxl import Workbook
import random
from scipy.stats import rankdata
import statistics as stats

wb = Workbook()

ws = wb.active

ws['A1'] = '姓名'
ws['B1'] = '國文'
ws['C1'] = '英文'
ws['D1'] = '數學'
ws['E1'] = '自然'
ws['F1'] = '社會'
ws['G1'] = '總分'
ws['H1'] = '平均'
ws['I1'] = '名次'

namelist = 'ABCDEFGHIJ'

for i in namelist:
    t = []
    t.append(i)
    
    for j in range(5):
        g = random.randint(60,100)
        t.append(g)
        
    ws.append(t)

row_count = ws.max_row
for row in range(2 ,row_count + 1):
    t = []
    
    for column in 'BCDEF':
        cell_name = '{}{}'.format(column,row)
        t.append(ws[cell_name].value)

    total = sum(t)
    average = float(total)/len(t)
    
    ws['G' + str(row)] = total
    ws['H' + str(row)] = average

a = []
for row in range(2, len(namelist)+2):
    cell_name = '{}{}'.format('H',row)
    a.append(ws[cell_name].value)

r = rankdata(a, method='min')
#print(r)

new_r = []
for i in r:
    n = len(r)- i + 1
    new_r.append(n)
#print(new_r)

for row in range(2, len(namelist)+2):
    cell_name = '{}{}'.format('I',row)
    ws[cell_name] = new_r[row-2]

for column in 'BCDEFGH':
    p = []
    for row in range(2, len(namelist)+2):
        cell_name = '{}{}'.format(column,row)
        p.append(ws[cell_name].value)
        
    mean = stats.mean(p)
    std = stats.stdev(p)
    ws[column + str(12)] = mean 
    ws[column + str(13)] = std

ws['A12'] = '平均'
ws['A13'] = '標準差'
  
wb.save('test01.xlsx')
```
